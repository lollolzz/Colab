```python
"""
날짜 : 2021/10/18
이름 : 권능한
내용 : 파이썬 Matplotlib 실습하기

Maplotlib 라이브러리
- 파이썬 데이터 시각화 패키지
- Matplotlib 기반으로 하는 seabon 시각화 라이브러리

"""
import matplotlib.pyplot as plt
import numpy as np
```

```python

# 기본 차트
plt.title('plot1')
plt.plot([1,3,4,9,10,17])
plt.show
```

```python
# x축 설정
plt.title('plot2')
plt.plot([10, 20, 30, 40,  50, 60],[1,3,4,9,10,17])
plt.show
```

```python
# 차트 스타일
plt.title('plot3')
plt.plot([10, 20, 30, 40,  50, 60],[1,3,4,9,10,17], 'gs:')
plt.grid(True)
plt.show
```

```python
# 멀티 라인차트
values = np.arange(0, 5, 0.2)
print(values)

plt.title('plot4')
plt.plot(values, values, 'r--')
plt.plot(values, values ** 2, 'g--')
plt.plot(values, values ** 3, 'g--')
plt.show()
```

```python
# 막대 차트
plt.title('plot5')
plt.bar(['apple', 'banana', 'orange', 'grape'], [5,3,7,9])
plt.show()
```
